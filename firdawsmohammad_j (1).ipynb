if we observe the data, the columns reviews, price, installs, size, rating, are numericals but installs, price, size still have other type of values

```python
#for the installs column , it consists of '+', ',', ' ' string types after the numerical value, hence those string types are removed and installs column type is changed from object to int

data['Installs'] = data['Installs'].str.replace(',', '')
data['Installs'] = data['Installs'].str.replace('+', '')

# the Free string is replaced with numberical 0
data['Installs'] = data['Installs'].replace('Free', '0')

#converting to int type from string type
data['Installs'] = data['Installs'].astype(str).astype(int)
```

```python
# for the reviews column, there is only one type of string that is 3M which means 3 million, hence the value is changed to 3000000 and the object type is converted to int type

data[data['Reviews'].str.isnumeric() == False]
data['Reviews']  = data['Reviews'].str.replace('3.0M', '3000000')
data['Reviews'] = data['Reviews'].astype(str).astype(int)
```

```python
# the first line performs two operations at once: replacing "$" with an empty string and replacing "Everyone" with "0". The astype method is then used to cast the "Price" column to the float type.

# The second line uses the apply method to apply a lambda function that checks the value of each element in the "Price" column. If it's not equal to 0, the value remains unchanged. If it's equal to 0, it is set to 0.

data['Price'] = data['Price'].str.replace('$', '').str.replace('Everyone', '0').astype(float)
data.Price = data.Price.apply(lambda x: x if x != 0 else 0)
data.Price.head()
```

```python
from sklearn.preprocessing import LabelEncoder

obj = LabelEncoder()
data['Category'] = obj.fit_transform(data['Category'].astype('object'))

rating_labels = data['Content Rating'].unique()
rating_dict = {rating_labels[i]: i for i in range(len(rating_labels))}
data['Content Rating'] = data['Content Rating'].map(rating_dict).astype(int)

data.dtypes
```

# One Hot Encoding

One-hot encoding is a method used to represent categorical data in a numerical format by creating a binary column for each category and assigning a 1 or 0 to indicate the presence or absence of the category in a given row. In this case, we are using one-hot encoding for the categorical data of country, province, taster_name, and variety. This means that for each unique value in the categorical variable, a new binary column is created, and a 1 is placed in the corresponding column for the observation that has that value.

## Linear Regression

```python
from sklearn.metrics import r2_score, mean_squared_error 
from sklearn.linear_model import LinearRegression
from sklearn import datasets, linear_model
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train,y_train)

```

## Making predictions

```python
# Make predictions using the training set
y_pred = regr.predict(X_train)
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_train, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_train, y_pred))
r2 = r2_score(y_train,y_pred)
print('R^2 score on tarining set =',r2)
```

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))
r2 = r2_score(y_test,y_pred)
print('R^2 score on test set =',r2)
```

```python
# Make predictions using the validation set
y_pred = regr.predict(X_val)

# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_val, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_val, y_pred))
r2_val = r2_score(y_val,y_pred)
print('R^2 score on validation set =',r2_val)

```
